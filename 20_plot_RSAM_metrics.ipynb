```python
import obspy
from SAM import RSAM
SAM_DIR = '/data/OVSM/SAM'
startTime = obspy.core.UTCDateTime(2012,4,1)
endTime = obspy.core.UTCDateTime(2012,7,1)
rsamObj = RSAM.read(startTime, endTime, SAM_DIR=SAM_DIR, ext='csv')
rsamObj = rsamObj.select(component='Z')
rsamObj.plot(metrics='fratio', ylims=[-4, 4])
```
